# 04 — Sistema multiagente A2A-lite para Herdez Smart-Supply

Este notebook replantea el punto 04 para que el agente no sea solo un wrapper de Gemini, sino una arquitectura **multiagente con mentalidad A2A**.

La idea central es:

> XGBoost predice riesgo, el motor determinista calcula costo-beneficio y los agentes A2A-lite producen artefactos auditables de riesgo, costo, política y comunicación ejecutiva.

No estamos desplegando un servidor A2A real todavía. Para la demo usamos un **A2A-lite local**, que representa los conceptos importantes: `AgentCard`, `Task`, `Message` y `Artifact`. Esto mantiene el prototipo local-first y deja clara la evolución a producción.

## 1. Por qué replanteamos el 04

Antes teníamos un agente que tomaba el resultado del `03_decision_engine.py` y generaba un brief.

Ahora lo hacemos más profesional:

1. Definimos agentes con tarjetas de capacidad (`AgentCard`).
2. Cada agente recibe evidencia y produce un artefacto estructurado.
3. Usamos prompts con identidad, misión, metodología, límites y ejemplos.
4. Validamos salidas con Pydantic.
5. Tenemos fallback determinista para que la demo no falle si no hay API key.

Esto demuestra arquitectura, soft skill, creatividad y control técnico.

## 2. Conceptos usados de tus PDFs

De tus documentos tomamos cuatro ideas clave:

- La instrucción del agente debe definir objetivo, personalidad/perfil, restricciones y formato de salida.
- Las instrucciones profesionales se organizan con cinco patrones: identidad, misión, metodología, límites y ejemplos.
- Las respuestas estructuradas deben usar esquemas tipo Pydantic para producir JSON predecible.
- La temperatura debe ser baja en tareas fácticas o de decisión para reducir alucinaciones.

En este caso, por tratarse de costos, inventario y recomendaciones ejecutivas, usamos temperaturas bajas y límites estrictos: el agente **nunca inventa costos, unidades, SKUs ni probabilidades**.

## 3. Flujo A2A-lite

```text
A2ATask: responder a alertas de quiebre
   ↓
RiskLens → artifact: risk_analysis
   ↓
CostGuard → artifact: cost_analysis
   ↓
PolicyCritic → artifact: policy_critic
   ↓
ExecSupplyAI → artifact: executive_brief
```

Cada agente no recibe todo el mundo; recibe el contexto necesario y los artefactos anteriores. Eso hace el sistema más modular y auditable.

## 7. Cómo defenderlo en entrevista

Puedes decir:

> Replanteé el sistema de agentes usando una arquitectura A2A-lite local-first. Cada agente tiene una tarjeta de capacidades, recibe artefactos estructurados y produce un nuevo artefacto auditable. Esto me permite demostrar interoperabilidad y modularidad sin desplegar infraestructura cloud durante el prototipo. En producción, cada agente puede exponerse como servicio A2A real y ejecutarse en Cloud Run, Agent Platform o cualquier runtime compatible.

Y rematas con:

> Los agentes no reemplazan al decision engine: lo explican, lo critican y lo convierten en una recomendación ejecutiva. Los números críticos siguen viniendo del modelo y de funciones deterministas.
