# 05 Dashboard Streamlit explicado

Este notebook explica el último bloque del proyecto: el dashboard ejecutivo.

La idea no es solo mostrar gráficas, sino cerrar el caso de negocio: **riesgo → costo-beneficio → recomendación → explicación**.

## ¿Por qué necesitamos un dashboard?

El modelo XGBoost y el decision engine son potentes, pero el Director de Supply Chain necesita ver rápidamente:

- qué SKUs están en riesgo,
- en qué CEDIs ocurre el problema,
- cuánto dinero se podría perder,
- qué acción se recomienda,
- y por qué el agente sugiere esa acción.


## Diseño del dashboard

El dashboard se divide en 6 pestañas:

1. **Resumen ejecutivo**: KPIs y brief del agente.
2. **Alertas y recomendaciones**: tabla y gráficas operativas.
3. **Agentes A2A-lite**: registry, traza y artefactos.
4. **Modelo ML**: comparación de modelos y riesgo histórico.
5. **Simulador**: probar escenarios de transferencia.
6. **Arquitectura**: local-first y cloud-ready.


## Punto clave para entrevista

El dashboard demuestra soft skills porque traduce un sistema técnico complejo en una interfaz entendible para negocio.

Frase útil:

> “Diseñé el dashboard para que un director no tenga que entender XGBoost ni agentes; solo necesita ver el riesgo, el impacto económico y la acción recomendada.”


## Cómo correrlo

```bash
streamlit run src/05_streamlit_dashboard.py
```

Para Streamlit Cloud, el proyecto incluye un `app.py` en la raíz.

```text
Excel histórico
   ↓
pandas / feature engineering
   ↓
XGBoost
   ↓
Decision Engine determinista
   ↓
Agentes A2A-lite
   ↓
Streamlit Dashboard
```
